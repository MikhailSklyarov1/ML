```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
from PIL import Image
import joblib
```

```python
dataTest = pd.read_csv("test.csv", dtype=float)
dataTest.head()
```

```python
dataTest.count()
```

```python
dataTrain = pd.read_csv("train.csv", dtype=float)
dataTrain.head()
```

```python
dataTrain.count()
```

```python
dataArrayTrain = dataTrain.values
dataArrayTest = dataTest.values
```

```python
dataArrayTrain[:,1:] = dataArrayTrain[:,1:] / 255
dataArrayTest[:,1:] = dataArrayTest[:,1:] / 255
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(20,), activation='logistic', solver='sgd', verbose=True, random_state=1, max_iter=600, alpha=0.001, learning_rate_init=0.01)
```

```python
mlp.fit(dataArrayTrain[:,1:], dataArrayTrain[:,0])
```

```python
print("Training set score: %f" % mlp.score(dataArrayTrain[:,1:], dataArrayTrain[:,0]))
#print("Test set score: %f" % mlp.score(dataArrayTest[:,1:], dataArrayTest[:,0]))
```

```python
estimator = MLPClassifier(activation='logistic',solver='sgd', verbose=True)
parameters = {
    'learning_rate_init':[0.0001, 0.001, 0.01, 0.1, 1],
    'alpha':[0.0001, 0.001, 0.01, 0.1, 1],
    'hidden_layer_sizes':[(20,), (5,5,5), (30,), (10,10), (1,2,3,4)],
    'max_iter':[100, 500, 600, 700, 1000],
}

clf = GridSearchCV(estimator, parameters, cv = 4, n_jobs = -1, verbose=3)
```

```python
clf.fit(dataArrayTrain[:,1:], dataArrayTrain[:,0])
```

```python
print("Training set score: %f" % clf.score(dataArrayTrain[:,1:], dataArrayTrain[:,0]))
#print("Test set score: %f" % clf.score(dataArrayTest[:,1:], dataArrayTest[:,0]))
```
