```python
qa_model = pipeline("question-answering", model="AlexKay/xlm-roberta-large-qa-multilingual-finedtuned-ru")
```

```python
context="""Тут написан контекст, по которому нейросеть будет отвечать на вопросы. Мы учимся в учебном центре Юнит и изучаем машинное обучение.
            Здесть может быть много текста. 
         """
```

```python
question="Чем мы занимаемся?"
```

```python
results = qa_model(question = question, context = context)
```

```python
results
```
